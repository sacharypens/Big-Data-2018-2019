```python
results = ""
for province in root:
    for town in province:
        row = town.tag
        for tabel_uitslagen2012 in town.iter('tabel_uitslagen2012'):
            for actual_votes in tabel_uitslagen2012.iter('actual_votes'):
                row += "," + actual_votes.text
        for party in town.iter('lijst_-2_2012'):
            row += "," + party[2].text
            break
        else:
            row += "," + "null"
        print(row)
        results += row + "\n"
results = results[:-1]
print(results)
file = open("results_for_NVA_2012.csv", "w")
file.write(results)
```
