# Preparing data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "results_for_NVA_2012.csv"
data = np.recfromcsv(file, encoding=None)
print(data)
```

```python
no_nva = 0
total_votes = []
nva_votes = []
towns = []
for town in data:
    if(town[2] == -1):
        no_nva += 1
    else:
        nva_votes.append(town[2])
        total_votes.append(town[1])
        towns.append(town[0])
print("number of towns where nva was not present: " + str(no_nva))
print("total votes(" + str(len(total_votes)) + "):")
print(total_votes)
print("nva votes(" + str(len(nva_votes)) + "):")
print(nva_votes)
print("all towns where nva was present:")
print(towns)
```

```python
total_votes_np = np.array(total_votes)
nva_votes_np = np.array(nva_votes)
```

```python
percentage_votes_np = nva_votes_np / total_votes_np * 100
```

```python
for i in range(len(towns)):
    print(towns[i],":",percentage_votes_np[i])
```

# Statistics

## Centrummaten

```python
print("the average votes per town is " + str(nva_votes_np.mean()))
print("the avarage total votes per town is " + str(total_votes_np.mean()))
print("the avarage percentage of votes per town is " + str(percentage_votes_np.mean()))
```

```python
print("the median votes per town is " + str(np.median(nva_votes_np)))
print("the median total votes per town is " + str(np.median(total_votes_np)))
print("the median percentage of votes per town is " + str(np.median(percentage_votes_np)))
```
