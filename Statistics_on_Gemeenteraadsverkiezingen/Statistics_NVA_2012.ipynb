# Preparing data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "results_for_NVA_2012.csv"
data = pd.read_csv(file)
print(data)
```

```python
print(data[data["NVA"]> 0]["NVA"].sum())
data_where_NVA_participated = data["NVA"]> 0
```

```python
print("number of towns where nva was not present: " + str(data[data_where_NVA_participated].count()[1]))
print("total votes:" + str(data["Totaal"].sum()))
print("total of votes where nva participated:" + str(data[data_where_NVA_participated]["Totaal"].sum()))
print("nva votes:" + str(data[data_where_NVA_participated]["NVA"].sum()))
print("all towns where nva was present:")
print(data[data_where_NVA_participated]["Gemeente"])
```

```python
data["percentage"] = data[data_where_NVA_participated]["Totaal"] / data[data_where_NVA_participated]["NVA"]
```

```python
data
```

```python
data_participated = data[data_where_NVA_participated]
data_participated
```

# Statistics (with only data where nva participated)

## Centrummaten

```python
print("the average NVA votes per town is " + str(data_participated["NVA"].mean()))
print("the avarage total votes per town is " + str(data_participated["Totaal"].mean()))
print("the avarage percentage of votes per town is " + str(data_participated["percentage"].mean()))
```

```python
print("the median NVA votes per town is " + str(data_participated["NVA"].median()))
print("the median total votes per town is " + str(data_participated["Totaal"].median()))
print("the median percentage of votes per town is " + str(data_participated["percentage"].median()))
```

```python
print("unique numbers in NVA column: " + str(data_participated["NVA"].nunique()))
print("The modus for this dataset would not mean a lot because all/most values are different")
```
